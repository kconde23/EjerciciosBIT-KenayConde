**¿Cuál es la cadena de comida rápida que tiene, en promedio, los productos con mayor cantidad de calorías?**

```python
import pandas as pd
import numpy as np

fastfood = pd.read_csv(r"C:\Users\User\Downloads\FastFoodNutritionMenuV2.csv", encoding='utf-8')
fastfood.columns = fastfood.columns.str.replace('\n', '', regex=False)
fastfood.head()
```

```python
fastfood['Company'].unique()
```

```python
fastfood = fastfood[['Company', 'Item', 'Calories']]
fastfood['Calories'] = pd.to_numeric(fastfood['Calories'], errors='coerce')
fastfood = fastfood.dropna(subset=['Calories'])

media_calories = fastfood.groupby('Company')['Calories'].mean()

calories_array = np.array(media_calories)

media_np = np.mean(calories_array)
desvio_np = np.std(calories_array)

print("Media general de calorías:", media_np)
print("Desviación estándar:", desvio_np)
print(media_calories)
```

### CONCLUSION

**Burger King es la cadena que ofrece los productos con mayor contenido calórico promedio dentro del dataset analizado.**
